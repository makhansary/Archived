Having $x = ih$, $y = jh$ and $z = kh$ and $x_A=\alpha h$, $y_A=\beta h$ and $z_A = \gamma h$ (_$\alpha$, $\beta$ and $\gamma$ are not intgers neccessarily_):

\begin{equation}
F(r) \psi_{n}(r)= \epsilon_{n} \psi_{n}(r)
\end{equation}

\begin{equation}
F(x,y,z) = h(x,y,z) + \sum_{m=1}^{\frac{N}{2}} [2 J_m(x,y,z) - K_m(x,y,z)]
\end{equation}

**One Electron Part**

\begin{equation}
h(x,y,z) {\color{red} {\times \psi_{n}(r)}} = \left[(-\frac{1}{2}\nabla^{2}) + \sum_{A=1}^{M} (\frac{-Z_A}{r_{iA}})\right] \color{red}{\times \psi_{n}(r)}
\end{equation}

\begin{align}
    \psi_{i,j,k}: \quad & \frac{3}{h^2} + \frac{1}{h}[\sum_{A=1}^{M}(\frac{-Z_A}{\sqrt(i - \alpha)^2 + (j - \beta)^2 + (k - \gamma)^2})]\\
    \psi_{i-1,j,k}: \quad & -\frac{1}{2{h^2}} \\
    \psi_{i+1,j,k}: \quad & -\frac{1}{2{h^2}} \\
    \psi_{i,j-1,k}: \quad & -\frac{1}{2{h^2}} \\
    \psi_{i,j+1,k}: \quad & -\frac{1}{2{h^2}} \\
    \psi_{i,j,k-1}: \quad & -\frac{1}{2{h^2}} \\
    \psi_{i,j,k+1}: \quad & -\frac{1}{2{h^2}} \\
\end{align}

<font color="green">Appears on both diagonal and off-diagonal.</font>

**Coulomb Part**

\begin{equation}
J_m(r) {\color{red}{\times \psi_{n}(r)}} = \left[\int \psi_{m}^{*}(r) \frac{1}{r_{12}} \psi_{m}(r) dr \right] \color{red}{\times \psi_{n}(r)}
\end{equation}

\begin{equation}
J_m(x,y,z) = \int \psi_{m}^{*}(x,y,z) \frac{1}{r_{12}} \psi_{m}(x,y,z) dxdydz
\end{equation}

\begin{equation}
J_m(x,y,z) {\color{red}{\times \psi_{n}(x,y,z)}} = \left[ h^2 \sum_{si \neq i}^{N_i}\sum_{sj \neq j}^{N_j}\sum_{sk \neq k}^{N_k}  \frac{\psi_{{m}_{\color{blue}{si,sj,sk}}}^{*} \psi_{{m}_{\color{blue}{si,sj,sk}}}}{\sqrt((si - i)^2 + (sj - j)^2 + (sk - k)^2)} \right] \color{red}{\times \psi_{i,j,k}}
\end{equation}

<font color="green">Appears on diagonal only.</font>

\begin{equation}
\sum_{m=1}^{\frac{N}{2}} [2 J_m(x,y,z)] = 2 \times \sum_{m=1}^{\frac{N}{2}} \left( \left[ h^2 \sum_{si \neq i}^{N_i}\sum_{sj \neq j}^{N_j}\sum_{sk \neq k}^{N_k}  \frac{\psi_{{m}_{\color{blue}{si,sj,sk}}}^{*} \psi_{{m}_{\color{blue}{si,sj,sk}}}}{\sqrt((si - i)^2 + (sj - j)^2 + (sk - k)^2)} \right] \right)
\end{equation}

**Exchange Part**

\begin{equation}
K_m(r) {\color{red}{\times \psi_{n}(r)}} = \left[\int \psi_{m}^{*}(r) \frac{1}{r_{12}} \psi_{n}(r) dr \right] \color{red}{\times \psi_{m}(r)}
\end{equation}

\begin{equation}
K_m(x,y,z) {\color{red}{\times \psi_{n}(x,y,z)}} = \left[ \int \psi_{m}^{*}(x,y,z) \frac{1}{r_{12}} \psi_{n}(x,y,z) dxdydz \right] \color{red}{\times \psi_{m}(x,y,z)}
\end{equation}

\begin{equation}
K_m(x,y,z) {\color{red}{\times \psi_{n}(x,y,z)}} = \left[ h^2 \sum_{si \neq i}^{N_i}\sum_{sj \neq j}^{N_j}\sum_{sk \neq k}^{N_k}  \frac{\psi_{{m}_{\color{blue}{si,sj,sk}}}^{*} \psi_{{n}_{\color{blue}{si,sj,sk}}}}{\sqrt((si - i)^2 + (sj - j)^2 + (sk - k)^2)} \right] \color{red}{\psi_{{m}_{i,j,k}}}
\end{equation}

\begin{equation}
K_m(x,y,z) {\color{red}{\times \psi_{n}(x,y,z)}} = \left[ h^2 \sum_{si \neq i}^{N_i}\sum_{sj \neq j}^{N_j}\sum_{sk \neq k}^{N_k}  \frac{\psi_{{m}_{\color{blue}{si,sj,sk}}}^{*} \color{red}{\psi_{{m}_{i,j,k}}} \psi_{{n}_{\color{blue}{si,sj,sk}}}}{\sqrt((si - i)^2 + (sj - j)^2 + (sk - k)^2)} \right]
\end{equation}

\begin{equation}
K_m(x,y,z) {\color{red}{\times \psi_{n}(x,y,z)}} = \left[ h^2 \sum_{si \neq i}^{N_i}\sum_{sj \neq j}^{N_j}\sum_{sk \neq k}^{N_k}\frac{\psi_{{m}_{\color{blue}{si,sj,sk}}}^{*} \color{red}{\psi_{{m}_{i,j,k}}} }{\sqrt((si - i)^2 + (sj - j)^2 + (sk - k)^2)} \psi_{{n}_{\color{blue}{si,sj,sk}}} \right]
\end{equation}

<font color="green">Appears on every row eveywhere expect on diagonal, each element is:</font>

\begin{equation}
h^2  \left[ \sum_{si \neq i}^{N_i}\sum_{sj \neq j}^{N_j}\sum_{sk \neq k}^{N_k}\frac{\psi_{{m}_{\color{blue}{si,sj,sk}}}^{*} \color{red}{\psi_{{m}_{i,j,k}}} }{\sqrt((si - i)^2 + (sj - j)^2 + (sk - k)^2)} \right]
\end{equation}

\begin{equation}
\sum_{m=1}^{\frac{N}{2}} [- K_m(x,y,z)] = - h^2 \sum_{m=1}^{\frac{N}{2}} \left[  \sum_{si \neq i}^{N_i}\sum_{sj \neq j}^{N_j}\sum_{sk \neq k}^{N_k}\frac{\psi_{{m}_{\color{blue}{si,sj,sk}}}^{*} \color{red}{\psi_{{m}_{i,j,k}}} }{\sqrt((si - i)^2 + (sj - j)^2 + (sk - k)^2)} \right]
\end{equation}

In matrix form $B \times \psi$:

The column vector of $\psi$ is:

\begin{equation}
\psi_{n} =
\begin{bmatrix}
\psi_{1,1,1} \\
\psi_{1,1,2} \\
\vdots \\
\psi_{1,1,K-1} \\
\psi_{1,1,K} \\
\psi_{1,2,1} \\
\psi_{1,2,2} \\
\vdots \\
\psi_{1,2,K-1} \\
\psi_{1,2,K} \\
\psi_{1,K-1,1} \\
\psi_{1,K-1,2} \\
\vdots \\
\psi_{1,K-1,K-1} \\
\psi_{1,K-1,K} \\
\psi_{1,K,1} \\
\psi_{1,K,2} \\
\vdots \\
\psi_{1,K,K-1} \\
\psi_{1,K,K} \\
\vdots \\
\vdots \\
\psi_{K-1,K,K} \\
\psi_{K,K,K}
\end{bmatrix}
\end{equation}

Each row in matrix $B \left[K^3, K^3 \right]$ follows the pattern:

<small>
\begin{equation}
\small
\begin{bmatrix}
(1,1,1) & (1,1,2) & \cdots & (1,1,K-1) & (1,1,K) & (1,2,1) & (1,2,2) & \cdots & (1,2,K-1) & (1,2,K) & (1,K-1,1) & (1,K-1,2) & \cdots & (1,K-1,K-1) & (1,K-1,K) & (1,K,1) & (1,K,2) & \cdots & (1,K,K-1) & (1,K,K) & \cdots & \cdots & (K-1,K,K) & (K,K,K) \\
\end{bmatrix}
\end{equation}
</small>

```python
import numpy as np
from scipy.linalg import eigh
import matplotlib.pyplot as plt
from tqdm import tqdm
from mpl_toolkits.mplot3d import Axes3D
from sympy import symbols, sqrt, solve
```

```python
def build_B(K, h, M, PSI): # builds matrix B in system: B*Psi
  n = K**3 # matrix size [n,n]
  total_iterations = K ** 3
  bar_B = tqdm(total=total_iterations, desc='Building HF operator', \
               position=0, leave=True)
  B = np.zeros((n, n)) # memory preallocation
  for i in range(1,K+1): # rolling to create row per equation
      for j in range(1,K+1): # rolling to create row per equation
          for k in range(1,K+1): # rolling to create row per equation
              bar_B.update(1)
              index = (i-1) * K**2 + (j-1) * K + (k-1) # index: row per equation
              B[index, index] += 3/(h*h) # contributions from one electron
              if  M.size != 0: # 1 = nuclei inlcuded, 0 = electronic only
                elenuc = 0
                for m in range(1,M.shape[0]+1):
                  elenuc += ((-M[m-1,-1]/np.sqrt((i - M[m-1,0])**2 + \
                   (j - M[m-1,1])**2 + (k - M[m-1,2])**2))/h) \
                   # contributions from electron-nuclei
              else:
                elenuc = 0
              #
              B[index, index] += elenuc
              #
              for mm in range(1,int(ne/2)+1):
                coul = 0
                for ii in range(1,K+1): # bringing in Coulomb
                      for jj in range(1,K+1):
                          for kk in range(1,K+1):
                              indexx = (ii-1) * K**2 + (jj-1) * K + (kk-1)
                              if not (ii == i and jj == j and kk == k):
                                coul += (((h**2)/np.sqrt((ii - i)**2 + \
                                 (jj - j)**2 + (kk - k)**2))*\
                                  (PSI[indexx, mm-1]\
                                   *np.conjugate(PSI[indexx, mm-1]))) \
                                   # contributions from Coulomb
              #
              B[index, index] += (2*coul) #  Coulomb
              #
              for mm in range(1,int(ne/2)+1):
                for ii in range(1,K+1): # bringing in exchange
                      for jj in range(1,K+1):
                          for kk in range(1,K+1):
                              indexx = (ii-1) * K**2 + (jj-1) * K + (kk-1)
                              if not (ii == i and jj == j and kk == k):
                                exch = (-(((h**2)/np.sqrt((ii - i)**2 +\
                                                          (jj - j)**2 + \
                                                           (kk - k)**2))\
                                          *(PSI[index, mm-1]*\
                                            np.conjugate(PSI[indexx, mm-1]))))\
                                             # exchange
                                B[index, indexx] += exch #exchange
              #
              if i > 1:
                  B[index, index - K**2] += -0.5/(h*h) # one electron
              if i < K:
                  B[index, index + K**2] += -0.5/(h*h) # one electron
              if j > 1:
                  B[index, index - K] += -0.5/(h*h) # one electron
              if j < K:
                  B[index, index + K] += -0.5/(h*h) # one electron
              if k > 1:
                  B[index, index - 1] += -0.5/(h*h) # one electron
              if k < K:
                  B[index, index + 1] += -0.5/(h*h) # one electron
  return B
```

```python
def compute_nhn(ne, K, PSI): # computes nhn contributions in E0
  nhn = 0
  bar_nhn = tqdm(total=(ne * (K ** 3)), desc='computing energy : task 1/2', \
                 position=0, leave=True)
  for m in range(1,ne+1):
    for i in range(1,K+1):
          for j in range(1,K+1):
              for k in range(1,K+1):
                  index = (i-1) * K**2 + (j-1) * K + (k-1)
                  bar_nhn.update(1)
                  #h1 - Div
                  Div = 0
                  if (index + K**2 < K**3 and index - K**2 > 0): # inner pointsX
                    Div = PSI[index, m-1] + \
                     (PSI[index - K**2, m-1] + \
                                             PSI[index + K**2, m-1] + \
                                             PSI[index - K, m-1]+ \
                                             PSI[index + K, m-1] + \
                                             PSI[index - 1, m-1] + \
                                             PSI[index + 1, m-1])/6
                  #
                  if (index + K < K**3 and index - K > 0): # inner pointsY
                    Div = PSI[index, m-1] + (PSI[index - K, m-1]+ \
                                             PSI[index + K, m-1] + \
                                             PSI[index - 1, m-1] + \
                                             PSI[index + 1, m-1])/4
                  #
                  if (index + 1 < K**3 and index - 1 > 0): # inner pointsZ
                    Div = PSI[index, m-1] + (PSI[index - 1, m-1] + \
                                             PSI[index + 1, m-1])/2
                  #h1 - ZA
                  DUM = 0
                  for mm in range(1,M.shape[0]+1): # nuclei
                    if not ((i == M[mm-1,0]) \
                        and (j == M[mm-1,1]) \
                        and (k == M[mm-1,2])):
                      DUM += ((-M[mm-1,-1]/\
                             np.sqrt((i - M[mm-1,0])**2 +\
                              (j - M[mm-1,1])**2 + \
                               (k - M[mm-1,2])**2))/h)
                  # h1
                  h1 = Div + DUM
                  #
                  nhn += (np.conjugate(PSI[index, m-1])*PSI[index, m-1]*(h**3)\
                          *(h1/(h**2)))
  return nhn
```

```python
def compute_nnmmOnmmn(ne, K, PSI): # computes nnnmm+nmmn contributions in E0
  nnmm = 0
  nmmn = 0
  bar_nnmmOnmmn = tqdm(total=(ne * ne * (K ** 3)), \
                       desc='computing energy : task 2/2', position=0, \
                       leave=True)
  for n in range(1,ne+1): # sum
    for m in range(1,ne+1): # sum
    # point 1
      for i1 in range(1,K+1): #
            for j1 in range(1,K+1):
                for k1 in range(1,K+1):
                    index1 = (i1-1) * K**2 + (j1-1) * K + (k1-1)
                    # point 2
                    for i2 in range(1,K+1):
                          for j2 in range(1,K+1):
                              for k2 in range(1,K+1):
                                  index2 = (i2-1) * K**2 + (j2-1) * K + (k2-1)
                                  bar_nnmmOnmmn.update(1)
                                  if ((i1 != i2) and (j1 != j2) and (k1 != k2)):
                                    r12 = 1/np.sqrt((i1 - i2)**2+\
                                     (j1 - j2)**2+(k1 - k2)**2)
                                    # nnmm
                                    nnmm += np.conjugate(PSI[index1, n-1])*\
                                    PSI[index2, m-1]*r12*PSI[index1, n-1]*\
                                    PSI[index2, m-1]*(h**3)*(h**3)
                                    # nmmn
                                    nmmn += -(np.conjugate(PSI[index1, n-1])*\
                                    PSI[index2, m-1]*r12*PSI[index2, n-1]*\
                                    PSI[index1, m-1]*(h**3)*(h**3))
  nnmmOnmmn = 0.5*(nnmm + nmmn)
  return nnmmOnmmn
```

```python
def energy(ne, K, PSI): # system energy E0
  E0 = compute_nhn(ne, K, PSI) + compute_nnmmOnmmn(ne, K, PSI)
  return E0
```

```python
def calculate(flag, L, h, ZA, ne, M, maxIter): # solve eigenvalue problem
  energy_store = []
  # initializing
  bar_I = tqdm(total=maxIter, desc='Iteration', position=0, leave=True)
  PSI=initialize_psi(K, ne, L)
  print('\n +++ computing guess set energy (E0) +++ ')
  E0 = energy(ne, K, PSI)
  energy_store.append(E0)
  print('\n guess set energy (E0):' f'{E0}')
  # setup iterations
  DIFF = 10000
  iter = 1
  while (iter < maxIter):
    bar_I.update(iter)
    if iter > 1:
      if DIFF < 1e-3:
        print('\n met incremental change after ' f'{iter}' ' iterations')
        break
      else:
        E0 = Ei
        energy_store.append(E0)
        PSI = eigenvectors[:,0:ne]
    B = build_B(K, h, M, PSI) # matrix B in system B*psi
    eigenvalues, eigenvectors = eigh(B)
    print('+++ computing current set energy (E0) +++ ')
    Ei = energy(ne, K, eigenvectors[:,0:ne])
    print('\n current set energy (E0):' f'{Ei}')
    if flag == 1:  #  ------------------------- with energy minimization
      DIFF = 100*np.abs((E0 - Ei)/E0)
    if flag == 0:  #  ------------------------- without energy minimization
      DIFF = np.sum(np.abs(PSI[:,0] - eigenvectors[:,0]))
    iter += 1
  return energy_store, eigenvalues, eigenvectors
```

_Candidate initial set of individually orthonormal functions $\psi$_

For a safe start, a 1D individually orthonormal set can be built with $(n = 1, 2, 3, \ldots, N)$, and $L$ as interval length:

\begin{align*}
\psi_n(s) = \begin{cases}
\sqrt{\frac{2}{L}} \cdot \cos\left(\frac{n\pi s}{L}\right) & \text{if } n \text{ is even} \\
\sqrt{\frac{2}{L}} \cdot \sin\left(\frac{n\pi s}{L}\right) & \text{if } n \text{ is odd}
\end{cases}
\end{align*}

and then use, $\psi_n(x,y,z) = \psi_n(x) \cdot \psi_n(y) \cdot \psi_n(z)$.

```python
def initialize_psi(K, ne, L): # builds an initial guess of orthonormal set psi
  PSI = np.zeros((K**3,ne)) # memory preallocation
  iX, jY, kZ = local_map(K)
  for vfun in range(1, ne+1):
    if vfun % 2 == 0: # even
      PSI[:,vfun-1] = np.sqrt(2/L)*np.cos(vfun*np.pi*iX/L) \
      * np.sqrt(2/L)*np.cos(vfun*np.pi*jY/L) \
      * np.sqrt(2/L)*np.cos(vfun*np.pi*kZ/L)
    else: # odd
      PSI[:,vfun-1] = np.sqrt(2/L)*np.sin(vfun*np.pi*iX/L) \
      * np.sqrt(2/L)*np.sin(vfun*np.pi*jY/L) \
      * np.sqrt(2/L)*np.sin(vfun*np.pi*kZ/L)
  return PSI
```

```python
def local_map(K): # offloaded mapping of index with 3D real coordinates
  ix = []
  jy = []
  kz = []
  for i in range(1,K+1):
      for j in range(1,K+1):
          for k in range(1,K+1):
            ix.append(i)
            jy.append(j)
            kz.append(k)
  iX = np.array(ix)
  jY = np.array(jy)
  kZ = np.array(kz)
  return iX, jY, kZ
```

```python
def plot_mesh(K): # visualizes the built mesh / K-points
  iX, jY, kZ = local_map(K)
  fig = plt.figure(figsize=(5.5, 3.5))
  ax = fig.add_subplot(121, projection='3d')
  ax.scatter(iX, jY, kZ, c='black', alpha=0.5, s=2)
  ax.set_xlabel('X axis')
  ax.set_ylabel('Y axis')
  ax.set_zlabel('Z axis')
  ax.view_init(elev=30, azim=45)
  ax.set_title('3D Grid Points')
  # XY view
  ax_xy = fig.add_subplot(222)
  ax_xy.scatter(iX, jY, c='black', alpha=0.5, s=2)
  ax_xy.set_xlabel('X axis')
  ax_xy.set_ylabel('Y axis')
  ax_xy.set_title('XY view')
  # XZ view
  ax_xz = fig.add_subplot(224)
  ax_xz.scatter(iX, kZ, c='black', alpha=0.5, s=2)
  ax_xz.set_xlabel('X axis')
  ax_xz.set_ylabel('Z axis')
  ax_xz.set_title('XZ view')
  plt.tight_layout()
  plt.show()
  plt.close()
```

```python
def plot_eigenvalues(eigenvalues, ne): # visualize all and first -ne-
  plt.figure(figsize=(7.5,2.5))
  plt.plot(eigenvalues, '-k', label='all eigenvalues')
  plt.plot(eigenvalues[0:ne], '+r', label='first ' f'{ne}' ' eigenvalues')
  plt.xlabel('one electron function index')
  plt.ylabel('eigenvalues')
  plt.legend()
  plt.show()
  plt.close()
```

```python
def plot_eigenvectors(eigenvectors, ne): # visualize all and first -ne-
  plt.figure(figsize=(7.5,2.5))
  for col in range(eigenvectors.shape[1]):
    plt.plot(eigenvectors[:,col])
  plt.xlabel('index')
  plt.ylabel('eigenvectors')
  plt.title('all eigenvectors')
  plt.show()
  plt.close()
  plt.figure(figsize=(7.5,2.5))
  for col in range(ne):
    plt.plot(eigenvectors[:,col])
  plt.xlabel('index')
  plt.ylabel('eigenvectors')
  plt.title('first ' f'{ne}' ' eigenvectors')
  plt.show()
  plt.close()
```

```python
def plot_sai3D(K, ne, eigenvectors, eigenvalues): # visualize all and first -ne-
#                                                 in spherical
  # pulling and plotting the first -ne- eigenvectors/values
  iX, jY, kZ = local_map(K)
  r = np.sqrt(iX**2 + jY**2 + kZ**2)
  theta = np.arccos(kZ / r)
  phi = np.arctan2(jY, iX)
  #
  for i in range(ne):
    fig = plt.figure(figsize=(9.5, 4.5))
    ax = fig.add_subplot(121, projection='3d')
    ax.scatter(phi, theta, eigenvectors[:,i], c=eigenvectors[:,i], \
               cmap='viridis', alpha=0.5, s=2)
    ax.set_xlabel(r'$\phi$')
    ax.set_ylabel(r'$\theta$')
    ax.set_zlabel(r'$\psi$')
    #ax.view_init(elev=30, azim=45)
    ax.set_title(fr'$\psi_{i+1}$ at eigenvalue: {eigenvalues[i]:.2f}')
    # XY view
    ax_xy = fig.add_subplot(222)
    ax_xy.scatter(phi, eigenvectors[:,i], c='black', alpha=0.5, s=2)
    ax_xy.set_xlabel(r'$\phi$')
    ax_xy.set_ylabel(r'$\psi$')
    ax_xy.set_title(r'$\psi$' ' vs. 'r'$\phi$')
    # XZ view
    ax_xz = fig.add_subplot(224)
    ax_xz.scatter(theta, eigenvectors[:,i], c='black', alpha=0.5, s=2)
    ax_xz.set_xlabel(r'$\theta$')
    ax_xz.set_ylabel(r'$\psi$')
    ax_xz.set_title(r'$\psi$' ' vs. 'r'$\theta$')
    plt.tight_layout()
    plt.show()
    plt.close()
```
