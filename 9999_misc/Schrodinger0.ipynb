The (many body) SchrÃ¶dinger equation derived in course of Lectures - for the Wavefunction $\psi$ as $\psi(r_{i}^{\rightarrow},  R_{A}^{\rightarrow})$   where $r$ and $R$ being the position vector of electrons $i$ and nuclei $A$ ignoring the spin - is given as:

\begin{equation}
H\psi = E\psi
\end{equation}

where, the Hamiltonian $H$ was derived as:

\begin{equation}
H=\sum_{i=1}^{N}(-\frac{1}{2}\nabla^{2}_{i}) + \sum_{A=1}^{M}(-\frac{1}{2m_{A}}\nabla^{2}_{A}) + \sum_{i=1}^{N}\sum_{j>i}^{N}(\frac{1}{r_{ij}}) + \sum_{A=1}^{M}\sum_{A>B}^{M}(\frac{Z_A Z_B}{R_{AB}}) + \sum_{i=1}^{N}\sum_{A=1}^{M}(\frac{- Z_A}{r_{iA}})
\end{equation}

Using the **Born Oppenheimer approximation** the kinetic energy coming from the nuclei were neglected and only kinetic portion coming from the electrons were inlcuded. Therefore, the originl wavefunction were rewritten in an electronic part that depends on electronic coordinates as variables but nuclei as paramter, and a nuclear part that only depends on nuclei coordinates as variables, i.e.:

\begin{equation}
\psi = \psi_{elec}(r_{i}^{\rightarrow},  R_{A}^{\rightarrow}) \psi_{nucl}(R_{A}^{\rightarrow})
\end{equation}

After applying the Hamiltonian on the **_decomposed wavefunction_** and some mathematical manupilation, specially for the second derivative with repect to nuclear coordinates $R$ (as the other have the $\psi_{elec}(r_{i}^{\rightarrow},  R_{A}^{\rightarrow}) \psi_{nucl}(R_{A}^{\rightarrow})$ explicitely on both sides when looking at $H\psi = E\psi$), we got:

\begin{equation}
( \sum_{A=1}^{M}(-\frac{1}{2m_{A}}\nabla^{2}_{A}) + \sum_{A=1}^{M}\sum_{A>B}^{M}(\frac{Z_A Z_B}{R_{AB}}) + E_{elec}) \psi_{nucl} = E \psi_{nucl}
\end{equation}

where $E_{elec}$ is electronic part that is the average potential coming from electrons and felt by nuclei:

\begin{equation}
E_{elec}=\sum_{i=1}^{N}(-\frac{1}{2}\nabla^{2}_{i}) + \sum_{i=1}^{N}\sum_{j>i}^{N}(\frac{1}{r_{ij}}) + \sum_{i=1}^{N}\sum_{A=1}^{M}(\frac{- Z_A}{r_{iA}})
\end{equation}

> An iterative dynamics starts with $E_{elec}$ at current nuclei coordinates and then the $\psi_{nucl}$ is solved to produce the new nuclei coordinates.

For electronics-only problem, we could use:

\begin{equation}
H_{elec} \psi_{elec}= [\sum_{i=1}^{N}(-\frac{1}{2}\nabla^{2}_{i}) + \sum_{i=1}^{N}\sum_{j>i}^{N}(\frac{1}{r_{ij}}) + \sum_{i=1}^{N}\sum_{A=1}^{M}(\frac{- Z_A}{r_{iA}})] \psi_{elec} = E \psi_{elec}
\end{equation}

As we are dealing with one electron (no summation of any sort) in a well (no nuclei), then we could drop the last two terms and get:

\begin{equation}
H_{elec} \psi_{elec}= -\frac{1}{2}\nabla^{2} \psi_{elec} = E \psi_{elec}
\end{equation}

The problem is one dimensional, so we could write say in x direction:

\begin{equation}
-\frac{1}{2} \frac{{d^2}\psi_{elec}}{dx^2}  = E \psi_{elec}
\end{equation}

Given the form of this differential equation $-ay ''=by$, we could propose the general solution as:

\begin{equation}
\psi_{elec} = C_{1} Sin(\alpha x) + C_{2} Cos(\alpha x)
\end{equation}
where $\alpha$ is $\sqrt{\frac {b}{a}} = \sqrt{2E}$.

We use the boundary conditions arising from the well definition stating an inf within the well width require wavefunction to be 0 at the boundaries of the well:  

\begin{matrix}
\psi(x=0) = 0\\
\psi(x=L) = 0
\end{matrix}

Applying the boundary conditions, we get:

\begin{matrix}
C_{2} = 0\\
Sin(kL) = 0
\end{matrix}

From $Sin(\alpha L) = 0$ one gets $\alpha L = n\pi$. Therefore, we could rewrite:

\begin{equation}
E_n = \frac {n^{2} \pi^{2}} {2L^{2}}
\end{equation}

> $E_n$ are the allowed energy levels.

The wavefunctions are therefore:

\begin{equation}
\psi_{elec, n} = C_{1} Sin(n\pi \frac {x}{L})  
\end{equation}

The last step is to determine $C_1$, which can be done knowing that total probability of finding the electron within the well must be equal to 1. Therefore, we could write:

\begin{equation}
\int_{0}^{L} |\psi_{elec, n}|^2 dx = \int_{0}^{L} |C_{1} Sin(n\pi \frac {x}{L})|^2 dx= 1
\end{equation}

Therefore:

\begin{equation}
\int_{0}^{L} |C_{1}|^2 Sin^2(n\pi \frac {x}{L}) dx= 1 \\
\int_{0}^{L} |C_{1}|^2 [\frac {1}{2} - \frac{1}{2} cos(2n\pi \frac {x}{L})] dx= 1 \\
\frac {|C_{1}|^{2}}{2} L = 1 \\
C_{1} = \sqrt {\frac {2}{L}}
\end{equation}

Finally, the wavefunction is given as:

\begin{equation}
\psi_{elec, n} = \sqrt {\frac {2}{L}} Sin(n\pi \frac {x}{L})  
\end{equation}

The eigenvalues for n = 1, ..., 6 would be:

\begin{matrix}
n = 1 & & & E_n = \frac {\pi^{2}} {200}\\
n = 2 & & & E_n = \frac {4 \pi^{2}} {200}\\
n = 3 & & & E_n = \frac {9 \pi^{2}} {200}\\
n = 4 & & & E_n = \frac {16 \pi^{2}} {200}\\
n = 5 & & & E_n = \frac {25 \pi^{2}} {200}\\
n = 6 & & & E_n = \frac {36 \pi^{2}} {200}
\end{matrix}

We now can plot the first six wavefunctions as follows:

```python
# importing modules for plotting and calculations
import numpy as np
import matplotlib.pyplot as plt
#
L = 10 # width of well
x = np.linspace(0, L, 1000) # x coordinate to compute wavefunctions
N = 6 # first six wavefunctions
plt.figure(figsize=(10, 6)) # plot container
plt.title('ANALYTICAL: Wave Functions in an Infinite Potential Well')
plt.xlabel('Position (atomic units)')
plt.ylabel('Wave Function Amplitude')
#
for n in range(N):  # looping over requested wavefunctions
  Sai = np.sqrt(2/L)*np.sin((n+1)*np.pi*x/L) # computing wavefunctions
  plt.plot(x, Sai, label='n='f'{n+1}') # plotting wavefunctions
plt.legend() # adding legends
```

The one dimensional Schrodinger equation in x direction was:

\begin{equation}
-\frac{1}{2} \frac{{d^2}\psi_{elec}}{dx^2}  = E \psi_{elec}
\end{equation}

By discretization, we get:

\begin{equation}
-\frac{1}{2} \frac{\psi_{i+1} - 2\psi_{i} + \psi_{i-1}}{\Delta x^2}  = E \psi_{i}
\end{equation}


We expect the Hamiltonian matrix having these properties:

- On the diagonal elements correspond to $\frac{\psi_{i}}{\Delta x^2}$ and so are $\frac{1}{\Delta x^2}$
- Off the diagonal elements but the immediate up and down diagonal correspond to $-\frac{1}{2}\frac{\psi_{i+1}}{\Delta x^2}$ and $-\frac{1}{2}\frac{\psi_{i-1}}{\Delta x^2}$ and so are $-\frac{1}{2}\frac{1}{\Delta x^2}$.

Now we implement the code to perform the computations:

```python
# loading modules for computations
import numpy as np
from scipy.linalg import eigh
#
N = 50  # Number of grid points - min 50 requested
L = 10   # Width of the well
#
m = 1.0    # electron mass
dx = L / (N + 1) # discretized x coordinate
H_diag = (1.0 / (dx**2)) * np.ones(N) # Hamiltonian - Diagonal
H_diag_off = (-0.5 / (dx**2)) * np.ones(N - 1) # Hamiltonian - Off Diagonal
H = np.diag(H_diag) + np.diag(H_diag_off, k=1) + np.diag(H_diag_off, k=-1) #
# first term in the diagonal itself, the second is up and the last is down.
eigenvalues, eigenvectors = eigh(H) # solving for eigenvalues
n = 6 # first six energy levels
print("Energy levels:")
for n, energy in enumerate(eigenvalues[:n-1]):
    print(f"E_{n+1} = {energy:.4f}")
```

> a multiplier of an eigenvector is an eigenvector itself. So we force the numerical eigenvectors to have the same **sing** and compare again with analytical solution.

```python
eigenvectors_ns = np.sign(eigenvectors[0]) * eigenvectors
```

> The eigen vectors in analytical solution are normalized as enforced by $C_{1}$ criteria, therefore the numerical eigenvectors must be normalized first as:

```python
eigenvectors_n = eigenvectors_ns
for n in range(6):
    psi_n = eigenvectors_n[:, n]
    normalization_factor = np.sqrt(np.sum(np.abs(psi_n)**2) * dx)
    eigenvectors_n[:, n] /= normalization_factor
```

```python
# importing modules for plotting and calculations
x = np.linspace(0, L, N) # x coordinate to compute wavefunctions
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6)) # plot container
plt.title('NUMERICAL: Wave Functions in an Infinite Potential Well')
plt.xlabel('Position (atomic units)')
plt.ylabel('Wave Function Amplitude')
for n in range(6):
  plt.plot(x, eigenvectors_n[:,n], label='n='f'{n+1}')
plt.legend() # adding legends
```

Comparing to the analytical solution:

```python
# importing modules for plotting and calculations
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#
N = 50  # Number of grid points - min 50 requested
L = 10 # width of well
x = np.linspace(0, L, N) # x coordinate to compute wavefunctions
n = 6 # first six wavefunctions
#
E_a = [] # containers
E_n = [] # containers
ERR = [] # containers
for n in range(6):  # looping over
  E_a.append(((n+1)*(n+1))*(np.pi*np.pi)/(2*(L*L))) # computing wavefunctions
  E_n.append(eigenvalues[n])
  ERR.append(np.sum(100*np.abs(E_a[-1]-E_n[-1])/E_a[-1])) # evaluating the absolute error
Table = pd.DataFrame({"Analytical": E_a, "Numerical": E_n, "Absolute Error (%)": ERR}) # populating in a table
Table.index = range(1, len(Table) + 1) # fixing the DataFrame index
Table = Table.rename_axis('n', axis=0) # calling index as n
Table
```
