A **_s_**-type (spherical symmetry) Guassian function centered at some position $\mu_A$ with a standard deviation of $\sigma$ (aka $\alpha$) can be given as:

\begin{equation}
G(r) = \frac{1}{\sqrt{2\pi}\sigma} e^{-\frac{(r-\mu_A)^2}{2\sigma^2}}
\end{equation}

In an explicit 3D spatial cartezian coordinate

\begin{equation}
G(x, y, z) = \frac{1}{(2\pi)^{3/2}\sigma_x\sigma_y\sigma_z} e^{-\frac{(x-\mu_x)^2}{2\sigma_x^2}-\frac{(y-\mu_y)^2}{2\sigma_y^2}-\frac{(z-\mu_z)^2}{2\sigma_z^2}}
\end{equation}

This version, by having three tunable paramters $\sigma_x$, $\sigma_y$, and $\sigma_z$, would adds more flexiblity for the base task in assignment [remaining on one single Gaussian function]. **These three paramters dictate the width of the orbitals** shown with $\beta$. By enforcing uniform deviation i.e. $\frac{1}{\alpha} = \sigma_x = \sigma_y = \sigma_z$, the more familiar form is, where $\beta = \frac{\alpha^2}{2}$:

\begin{equation}
G(x, y, z) = \frac{\alpha^3}{(2\pi)^{3/2}} \times e^ {-\beta \left[(x-\mu_x)^2+(y-\mu_y)^2+(z-\mu_z)^2 \right]}
\end{equation}

The atomic radius of Hydrogen ($r_H$) and Helium ($r_{He}$) atoms is **1.20** Angstroms and **1.40** Angstroms, respectively [[sourced from PubChemT](https://pubchem.ncbi.nlm.nih.gov/ptable/)].

\begin{equation}
FC = SC \epsilon
\end{equation}

where,

\begin{equation}
F_{ij} = h_{ij} + \sum^K_{l=1} \sum^K_{q=1} p_{ql} \left( \left[ij | lq\right]
 - \frac{1}{2} \left[iq | lj\right] \right)
\end{equation}

\begin{equation}
h_{ij} = \int \phi^*_i (\vec{r}) \left[ (-\frac{1}{2}\nabla^{2}) + \sum_{A=1}^{M} \left(\frac{-Z_A}{r_{iA}}\right) \right] \phi_j (\vec{r}) d \vec{r}
\end{equation}

\begin{equation}
\left[ij | lq\right] = \int \phi^*_i (\vec{r}_1) \phi_j (\vec{r}_1) \frac{1}{r_{12}} \phi^*_l (\vec{r}_2) \phi_q (\vec{r}_2) d \vec{r}_1 d \vec{r}_2
\end{equation}

\begin{equation}
\left[iq | lj\right] = \int \phi^*_i (\vec{r}_1) \phi_q (\vec{r}_1) \frac{1}{r_{12}} \phi^*_l (\vec{r}_2) \phi_j (\vec{r}_2) d \vec{r}_1 d \vec{r}_2
\end{equation}

\begin{equation}
p_{ql} = 2 \sum^{\frac{N}{2}}_{m=1} C_{qm} C^*_{lm}
\end{equation}

\begin{equation}
S_{ij} = \int \phi^*_i (\vec{r}) \phi_j (\vec{r}) d\vec{r}
\end{equation}

Neither of $h_{ij}$, $S_{ij}$, $\left[ij | lq\right]$ or $\left[iq | lj\right]$ change while iterating.

In **discretized** form on a unifrom gird with spacing **$h$**, where $x = uh$, $y = vh$ and $z = wh$ and $x_A=ah$, $y_A=bh$ and $z_A = ch$ (where _$a$, $b$ and $c$ are not intgers neccessarily_):

\begin{equation}
S_{ij} = h^3 \sum_{u=1}^{K} \sum_{v=1}^{K} \sum_{w=1}^{K} \phi^*_{i_{u,v,w}} \phi_{j_{u,v,w}}
\end{equation}

\begin{equation}
\left[ij | lq\right] =  h^5 \sum_{uO \neq uI}^{K}\sum_{vO \neq vI}^{K} \sum_{wO \neq wI}^{K} \sum_{uI = 1}^{K}\sum_{vI = 1}^{K} \sum_{wI = 1}^{K} \frac{\phi_{{i}_{uO,vO,wO}}^{*} \phi_{{j}_{uO,vO,wO}} \phi_{{l}_{uI,vI,wI}}^{*} \phi_{{q}_{uI,vI,wI}}}{\sqrt((uO - uI)^2 + (vO - vI)^2 + (wO - wI)^2)}
\end{equation}

\begin{equation}
\left[ iq | lj \right] =  h^5 \sum_{uO \neq uI}^{K}\sum_{vO \neq vI}^{K} \sum_{wO \neq wI}^{K} \sum_{uI = 1}^{K}\sum_{vI = 1}^{K} \sum_{wI = 1}^{K} \frac{\phi_{{i}_{uO,vO,wO}}^{*} \phi_{{q}_{uO,vO,wO}} \phi_{{l}_{uI,vI,wI}}^{*} \phi_{{j}_{uI,vI,wI}}}{\sqrt((uO - uI)^2 + (vO - vI)^2 + (wO - wI)^2)}
\end{equation}

\begin{equation}
h_{ij} = h^3 \sum_{u \neq a}^{K} \sum_{v \neq b}^{K} \sum_{w \neq c}^{K} \phi^*_{i_{u,v,w}} \left[ f + \frac{1}{h}[\sum_{A=1}^{M}(\frac{-Z_A}{\sqrt(u - a)^2 + (v - b)^2 + (w - c)^2})] \phi_{j_{u,v,w}} \right]
\end{equation}

with $f$ everwhere except at boundaries ($ u, v, w > 1$) being:

\begin{equation}
f = -\frac{1}{2h^2} \left[\phi_{j_{u+1,v,w}} + \phi_{j_{u-1,v,w}} + \phi_{j_{u,v+1,w}} + \phi_{j_{u,v-1,w}} + \phi_{j_{u,v,w+1}} + \phi_{j_{u,v,w-1}} - 6 \phi_{j_{u,v,w}} \right]
\end{equation}

at any combination of $u = 1$, $v = 1$ and/ $w = 1$, the respective terms and the factor preceeding $\phi_{j_{u,v,w}}$ drop.

```python
import numpy as np
import pandas as pd
from numpy import linalg as LA
from numpy.linalg import inv
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sympy import symbols, sqrt, solve
import time
```

```python
def Gaussian(K, Kg, M, h, ne, beta): # generates Gaussian basis function
  n = Kg*Kg*Kg # matrix dimensions
  Guass = np.zeros((n, K)) # memory allocation
  za = -1
  for gauss in range(K):
    za = za + 1
    if za >= M.shape[0]: # switching over nuclei
      za = 0
    #if gauss > M.shape[0]:
    #  O = (gauss+1)
    #else:
    #  O = 1
    alpha = np.sqrt(2*beta[gauss,0])
    for i in range(Kg):
      for j in range(Kg):
          for k in range(Kg):
            if (i!=M[za, 0] and j!=M[za, 1] and k!=M[za, 2]):
              index = i * K**2 + j * K + k # index
              Guass[index, gauss] = ((alpha**3)/((2*np.pi)**(3/2)))*\
              np.exp(-beta[gauss,0]*(h**2)*\
              ((i-M[za, 0])**2+(j-M[za, 1])**2+(k-M[za, 2])**2))
              if np.isnan(Guass[index, gauss]):
                Guass[index, gauss] = 0
  return Guass
```

```python
def calc_S(h, K, eigenvectors): # calculates S overlap matrix
  n = K # matrix dimensions
  S = np.zeros((n, n)) # memory allocation
  for wf1 in range(K): # pair 1
    for wf2 in range(K): # pair 2
      S[wf1, wf2] = np.matmul(np.conjugate(eigenvectors[:,wf1]), \
                                eigenvectors[:,wf2]) # vector multiplication
                                #                         instead of 3-for-loops
  return (h**3)*S
```

```python
def calc_ijOlq_iqOlj(h, K, eigenvectors, i, j, l, q): # [ij|lq] or [iq|lj] K=Kg
  # [ij|lq] if calc_ijOlq_iqOlj(h, K, eigenvectors, i, j, l, q)
  # [iq|lj] if calc_ijOlq_iqOlj(h, K, eigenvectors, i, q, l, j)
  SUM = 0
  for uO in range(K):
    for vO in range(K):
      for wO in range(K):
        indexO = uO * K**2 + vO * K + wO # index
        for uI in range(K):
          if uI != uO:
            for vI in range(K):
              if vI != vO:
                for wI in range(K):
                  indexI = uI * K**2 + vI * K + wI # index
                  if wI != wO:
                    dist = np.sqrt((uO-uI)**2+(vO-vI)**2+(wO-wI)**2)
                    dummy = np.conjugate(eigenvectors[indexO,i])*\
                    eigenvectors[indexO,j]*\
                    np.conjugate(eigenvectors[indexI,l])*eigenvectors[indexI,q]
                    SUM = SUM + dummy/dist
  return (h**5)*SUM
```

```python
def calc_hij(h, K, eigenvectors, M, i, j): # calculates hij - K=Kg
  SUM = 0
  f = 0
  for u in range(K):
    for v in range(K):
      for w in range(K):
        index = u * K**2 + v * K + w # index
        if ((u>=1 and u<K-1) and (v>=1 and v<K-1) and (w>=1 and w<K-1)):#!bound
          f = (-0.5/(h**2))*(eigenvectors[(u+1) * K**2 + v * K + w, j] +\
                             eigenvectors[(u-1) * K**2 + v * K + w, j] +\
                             eigenvectors[u * K**2 + (v+1) * K + w, j] +\
                             eigenvectors[u * K**2 + (v-1) * K + w, j] +\
                             eigenvectors[u * K**2 + v * K + (w+1), j] +\
                             eigenvectors[u * K**2 + v * K + (w-1), j] -\
                             6*eigenvectors[index, j])
        elif ((u<1 or u>=K-1) and (v>=1 and v<K-1) and (w>=1 and w<K-1)):
          f = (-0.5/(h**2))*(eigenvectors[u * K**2 + (v+1) * K + w, j] +\
                             eigenvectors[u * K**2 + (v-1) * K + w, j] +\
                             eigenvectors[u * K**2 + v * K + (w+1), j] +\
                             eigenvectors[u * K**2 + v * K + (w-1), j] -\
                             4*eigenvectors[index, j])
        elif ((u>=1 and u<K-1) and (v<1 or v>=K-1) and (w>=1 and w<K-1)):
          f = (-0.5/(h**2))*(eigenvectors[(u+1) * K**2 + v * K + w, j] +\
                             eigenvectors[(u-1) * K**2 + v * K + w, j] +\
                             eigenvectors[u * K**2 + v * K + (w+1), j] +\
                             eigenvectors[u * K**2 + v * K + (w-1), j] -\
                             4*eigenvectors[index, j])
        elif ((u>=1 and u<K-1) and (v>=1 and v<K-1) and (w<1 and w>=K-1)):
          f = (-0.5/(h**2))*(eigenvectors[(u+1) * K**2 + v * K + w, j] +\
                             eigenvectors[(u-1) * K**2 + v * K + w, j] +\
                             eigenvectors[u * K**2 + (v+1) * K + w, j] +\
                             eigenvectors[u * K**2 + (v-1) * K + w, j] -\
                             4*eigenvectors[index, j])
        if ((u<1 or u>=K-1) and (v<1 or v>=K-1) and (w>=1 and w<K-1)):
          f = (-0.5/(h**2))*(eigenvectors[u * K**2 + v * K + (w+1), j] +\
                             eigenvectors[u * K**2 + v * K + (w-1), j] -\
                             2*eigenvectors[index, j])
        if ((u<1 or u>=K-1) and (v>=1 and v<K-1) and (w<1 and w>=K-1)):
          f = (-0.5/(h**2))*(eigenvectors[u * K**2 + (v+1) * K + w, j] +\
                             eigenvectors[u * K**2 + (v-1) * K + w, j] -\
                             2*eigenvectors[index, j])
        if ((u>=1 and u<K-1) and (v<1 or v>=K-1) and (w<1 and w>=K-1)):
          f = (-0.5/(h**2))*(eigenvectors[(u+1) * K**2 + v * K + w, j] +\
                             eigenvectors[(u-1) * K**2 + v * K + w, j] -\
                             2*eigenvectors[index, j])
        else:
          f = 0
        SUMi = 0
        for za in range(M.shape[0]): # nuclei
          if (u!=M[za, 0] and v!=M[za, 1] and w!=M[za, 2]):
            dist = np.sqrt((u-M[za, 0])**2+(v-M[za, 1])**2+(w-M[za, 2])**2)
            dummy = -M[za, -1]/dist
          else:
            dummy = 0
          SUMi = SUMi + dummy
        SUM = SUM + np.conjugate(eigenvectors[index, i])*\
         (f + (1/h)*SUMi)*eigenvectors[index, j]
  return (h**3)*SUM
```

```python
def COEFF(K, ne): # initial guess for matrix C
  n=K
  C = np.random.rand(n,n)
  C[0:2*ne, :] = C[0:2*ne, :]*10 # optional: adding weight to first 2*ne
  return C
```

```python
def calc_pql(C, ne, q, l): # calculates p_ql
  SUM = 0
  for n in range(int(np.ceil(ne/2))):
    SUM = SUM + C[q, n]*np.conjugate(C[l, n])
  return 2*SUM
```

```python
def OneTimeOperations(K, Kg, M, h, PSI): # all one time operations
  n = K
  H = np.zeros((n,n))
  F2ndTerm = np.zeros((K*K,1)) # storing [ij|lq] - (1/2) [iq|lj]
  for wf1 in range(K): # pair 1
    for wf2 in range(K): # pair 2
      H[wf1, wf2] = calc_hij(h, K, PSI, M, wf1, wf2)
      dummy = 0
      for q in range(K):
        for l in range(K):
          index = q * K + l # index
          dummy = dummy + calc_ijOlq_iqOlj(h, Kg, PSI, wf1, wf2, l, q) - \
          0.5*calc_ijOlq_iqOlj(h, Kg, PSI, wf1, q, l, wf2)
      F2ndTerm[index, 0] = dummy
  S = calc_S(h, K, PSI)
  return H, F2ndTerm, S
```

```python
def energy(ne, K, Kg, PSI): # system energy E0
  E0 = compute_nhn(ne, K, Kg, PSI) + compute_nnmmOnmmn(ne, K, Kg, PSI)
  return E0
def compute_nnmmOnmmn(ne, K, Kg, PSI): # computes nnnmm+nmmn contributions in E0
  nnmm = 0
  nmmn = 0
  for n in range(K): # sum
    for m in range(K): # sum
    # point 1
      for i1 in range(Kg): #
            for j1 in range(Kg):
                for k1 in range(Kg):
                    index1 = i1 * Kg**2 + j1 * Kg + k1
                    # point 2
                    for i2 in range(Kg):
                          for j2 in range(Kg):
                              for k2 in range(Kg):
                                  index2 = i2 * Kg**2 + j2 * Kg + k2
                                  if ((i1 != i2) and (j1 != j2) and (k1 != k2)):
                                    r12 = 1/np.sqrt((i1 - i2)**2+\
                                    (j1 - j2)**2+(k1 - k2)**2)
                                    # nnmm
                                    nnmm += np.conjugate(PSI[index1, n])*\
                                    PSI[index2, m]*r12*PSI[index1, n]*\
                                    PSI[index2, m]*(h**3)*(h**3)
                                    # nmmn
                                    nmmn += -(np.conjugate(PSI[index1, n])*\
                                    PSI[index2, m]*r12*PSI[index2, n]*\
                                    PSI[index1, m]*(h**3)*(h**3))
  nnmmOnmmn = 0.5*(nnmm + nmmn)
  return nnmmOnmmn
def compute_nhn(ne, K, Kg, PSI): # computes nhn contributions in E0
  nhn = 0
  for m in range(ne):
    for i in range(Kg):
          for j in range(Kg):
              for k in range(Kg):
                  index = i * Kg**2 + j * Kg + k
                  #h1 - Div
                  Div = 0
                  if (index + Kg**2 < Kg**3 and index - Kg**2 > 0): # inner pointsX
                    Div = PSI[index, m] + \
                    (PSI[index - Kg**2, m] + \
                                            PSI[index + Kg**2, m] + \
                                            PSI[index - Kg, m]+ \
                                            PSI[index + K, m] + \
                                            PSI[index - 1, m] + \
                                            PSI[index + 1, m])/6
                  #
                  if (index + Kg < Kg**3 and index - Kg > 0): # inner pointsY
                    Div = PSI[index, m] + (PSI[index - Kg, m]+ \
                                            PSI[index + Kg, m] + \
                                            PSI[index - 1, m] + \
                                            PSI[index + 1, m])/4
                  #
                  if (index + 1 < Kg**3 and index - 1 > 0): # inner pointsZ
                    Div = PSI[index, m] + (PSI[index - 1, m] + \
                                            PSI[index + 1, m])/2
                  #h1 - ZA
                  DUM = 0
                  for mm in range(M.shape[0]): # nuclei
                    if not ((i == M[mm,0]) \
                        and (j == M[mm,1]) \
                        and (k == M[mm,2])):
                      DUM += ((-M[mm,-1]/\
                            np.sqrt((i - M[mm,0])**2 +\
                              (j - M[mm,1])**2 + \
                              (k - M[mm,2])**2))/h)
                  # h1
                  h1 = Div + DUM
                  #
                  nhn += (np.conjugate(PSI[index, m])*PSI[index, m]*(h**3)\
                          *(h1/(h**2)))
  return nhn
```

**Solving generalized eigenvalue problem via [arXiv:1903.11240 (2019)](https://arxiv.org/abs/1903.11240)**

```python
def eigen_decomposition(matrix): # from arXiv:1903.11240 (2019).
    eig_val, eig_vec = LA.eigh(matrix)
    idx = eig_val.argsort()#[::-1]
    eig_val = eig_val[idx]
    eig_vec = eig_vec[:, idx]
    Eigenvectors = eig_vec
    eigenvalues = eig_val
    eigenvalues = np.asarray(eigenvalues)
    Eigenvalues = np.diag(eigenvalues)
    return Eigenvectors, Eigenvalues
def Solve(A,B):
  Phi_B, Lambda_B = eigen_decomposition(matrix=B)
  lambda_B = Lambda_B.diagonal()
  a = lambda_B**0.5
  a = np.nan_to_num(a) + 0.0001
  # Lambda_B_squareRoot = np.diag(lambda_B**0.5)
  Lambda_B_squareRoot = np.diag(a)
  Phi_B_hat = Phi_B.dot(inv(Lambda_B_squareRoot))
  A_hat = (Phi_B_hat.T).dot(A).dot(Phi_B_hat)
  Phi_A, Lambda_A = eigen_decomposition(matrix=A_hat)
  Lambda = Lambda_A
  Phi = Phi_B_hat.dot(Phi_A)
  return Phi, Lambda
```

```python
def local_map(Kg): # offloaded mapping of index with 3D real coordinates
  ix = []
  jy = []
  kz = []
  for i in range(Kg):
      for j in range(Kg):
          for k in range(Kg):
            ix.append(i)
            jy.append(j)
            kz.append(k)
  iX = np.array(ix)
  jY = np.array(jy)
  kZ = np.array(kz)
  return iX, jY, kZ
```

```python
# Helium atom
K = 2 # number of basis
Kg = 8 # 3D grid points
rHE = 1.40
ss1, ss2 = 0.005*rHE, 0.75*rHE
beta = np.array([ss1, ss2])
beta = beta.reshape(2, 1)
L = (1+0.5)*2*rHE # extend of analysis 150% atom size
h = L / (Kg-1) # uniform spacing; x=ih, y=jh and z=kh
ZA = 2
ne = 2 # - number of electron functions in analysis, must be even(%2=0)
M = np.array([(L/2)/h, (L/2)/h, (L/2)/h, ZA])
M = M.reshape(1, 4)
maxIter = 100
PSI = Gaussian(K, Kg, M, h, ne, beta)
C0 = COEFF(K, ne)
start_time = time.time()
H, F2ndTermi, S = OneTimeOperations(K, Kg, M, h, PSI)
end_time = time.time()
execution_time = end_time - start_time
print("One time operation executation time:", execution_time, "seconds")
F = np.zeros((K,K)) # Fock matrix
for iter in range(maxIter):
  print('Iteration: ', f'{int(iter)}', ' / ', f'{int(maxIter)}')
  F2ndTerm = 0
  for q in range(K):
    for l in range(K):
      index = q * K + l # index
      F2ndTerm = F2ndTerm + calc_pql(C0, ne, q, l)*F2ndTermi[index,0]
  F = H + F2ndTerm
  if np.linalg.det(S) == 0: # if singular
    S = S + np.eye(S.shape[0]) * 1e-10
  #S_inv_sqrt = np.linalg.inv(np.sqrt(S))
  #F_prime = np.dot(np.dot(S_inv_sqrt, F), S_inv_sqrt)
  #eps, C_prime = np.linalg.eigh(F_prime)
  #C = np.dot(S_inv_sqrt, C_prime)
  C, eps = Solve(F,S) # from arXiv:1903.11240 (2019).
  Err = 100*np.sum(np.abs(C-C0)/C0)
  print(Err)
  if Err > 1e-6:
    C0 = C
  else:
    print('no significant change in C over two consecutive iterations')
    print('converged, exiting ...')
    break
  if iter > maxIter:
    print('max iteration exceeded')
    print('terminating ...')
    break
```
